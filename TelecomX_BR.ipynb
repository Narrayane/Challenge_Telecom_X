```python
df = pd.json_normalize(dados)
df
```

```python
df.info()
```

```python
# Verificando se hÃ¡ valores Ãºnicos em cada coluna.
for col in df.columns:
  print(f"valores Ãºnicos na coluna'{col}': {df[col].nunique()}")
  if df[col].nunique() < 50: # Imprima os valores apenas se nÃ£o forem muitos.
    print(df[col].unique())
    print('-' * 50)
```

```python
# Verificando se hÃ¡ valores duplicados:
print("Numero de duplicados", df.duplicated().sum())
```

```python
# Verifiancando se hÃ¡ valores nulos
print("Numero de Nulos", df.isnull().sum())
```

```python
# Verificando se hÃ¡ campos vazios ou em branco
df.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

```python
#Alterando a coluna "account.Charges.Total" para Float

df['account.Charges.Total'] = pd.to_numeric(df['account.Charges.Total'], errors='coerce')
print(df['account.Charges.Total'].dtype)
```

```python
# Eliminando as colunas de Churn

df = df[df['Churn'].str.strip() != '']
print("NÃºmero de linhas depois de eliminar as vazias de 'Churn':", len(df))
```

```python
# Reconferindo os vazios ou em branco apÃ³s o tratamento
df.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

```python
#Verificando e removendo se existir uma coluna chamada "Contas_diÃ¡rias" no dataframe.
df = df.drop('contas_diarias', errors='ignore')
```

```python
# Calculando os gastos diÃ¡rios
df['contas_diarias'] = ((df['account.Charges.Total'] / df['customer.tenure']) /30 ).round(2)

# Inserindo a coluna na posiÃ§Ã£o 19
df.insert(19, 'contas_diarias', df.pop('contas_diarias'))
```

#DistribuiÃ§Ã£o da EvasÃ£o

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import plotly.express as px
```

```python
# Saida de clientes 0 = contratos ativos 1 = contratos encerrados
percentual_saida_clientes = df['churn'].value_counts(normalize=True)*100
percentual_saida_clientes
```

```python
# Criando o grÃ¡fico de pizza
plt.figure(figsize=(6, 6))
plt.pie(percentual_saida_clientes, labels=['Permaneceram', 'SaÃ­ram'], autopct='%1.1f%%', colors=['green', 'red'], startangle=90, explode=(0.1, 0))

# Adicionar tÃ­tulo
plt.title('Taxa de EvasÃ£o de Clientes (Churn)', fontsize=14)

# Exibir o grÃ¡fico
plt.show()
```

```python
# EvasÃ£o por tempo de Contrato:
# agrupar tempo de contrato e evasÃ£o:
df_evasao_tempo_contrato = df.groupby(['tenure', 'churn']).size().unstack(fill_value=0)
# percentuual de evasÃ£o por tempo de contrato:
df_evasao_tempo_contrato['Cancelamentos (%)'] = (df_evasao_tempo_contrato[1] / df_evasao_tempo_contrato.sum(axis=1))*100
print(df_evasao_tempo_contrato[['Cancelamentos (%)']])
```

```python
fig = px.line(df_evasao_tempo_contrato.reset_index(),
              x='tenure',
              y='Cancelamentos (%)',
              markers=True,
              title = 'Taxa de evasÃ£o por meses de contrato')

fig.show()
```

Observou-se que a taxa de evasÃ£o Ã© mais alta entre clientes com menor tempo de contrato. A linha de evasÃ£o por tempo de contrato demonstra que a maior probabilidade de churn ocorre nos primeiros meses de contrato e tende a diminuir Ã  medida que o tempo de permanÃªncia aumenta.

## ConclusÃµes e RecomendaÃ§Ãµes

A anÃ¡lise aponta para os seguintes fatores como importantes contribuintes para a evasÃ£o de clientes na TelecomX:

1.  **Tempo de Contrato Curto:** Clientes nos primeiros meses de contrato sÃ£o mais propensos a cancelar.
