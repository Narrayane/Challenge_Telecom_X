```python
df = pd.json_normalize(dados)
df
```

```python
df.info()
```

```python
# Verificando se h√° valores √∫nicos em cada coluna.
for col in df.columns:
  print(f"valores √∫nicos na coluna'{col}': {df[col].nunique()}")
  if df[col].nunique() < 50: # Imprima os valores apenas se n√£o forem muitos.
    print(df[col].unique())
    print('-' * 50)
```

```python
# Verificando se h√° valores duplicados:
print("Numero de duplicados", df.duplicated().sum())
```

```python
# Verifiancando se h√° valores nulos
print("Numero de Nulos", df.isnull().sum())
```

```python
# Verificando se h√° campos vazios ou em branco
df.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

```python
#Alterando a coluna "account.Charges.Total" para Float

df['account.Charges.Total'] = pd.to_numeric(df['account.Charges.Total'], errors='coerce')
print(df['account.Charges.Total'].dtype)
```

```python
# Eliminando as colunas de Churn

df = df[df['Churn'].str.strip() != '']
print("N√∫mero de linhas depois de eliminar as vazias de 'Churn':", len(df))
```

```python
# Reconferindo os vazios ou em branco ap√≥s o tratamento
df.apply(lambda x: x.astype(str).str.strip() == '').sum()
```

```python
#Verificando e removendo se existir uma coluna chamada "Contas_di√°rias" no dataframe.
df = df.drop('contas_diarias', errors='ignore')
```

```python
# Calculando os gastos di√°rios
df['contas_diarias'] = ((df['account.Charges.Total'] / df['customer.tenure']) /30 ).round(2)

# Inserindo a coluna na posi√ß√£o 19
df.insert(19, 'contas_diarias', df.pop('contas_diarias'))
```

#Distribui√ß√£o da Evas√£o

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import plotly.express as px
```

```python
# Saida de clientes 0 = contratos ativos 1 = contratos encerrados
percentual_saida_clientes = df['churn'].value_counts(normalize=True)*100
percentual_saida_clientes
```

```python
# Criando o gr√°fico de pizza
plt.figure(figsize=(6, 6))
plt.pie(percentual_saida_clientes, labels=['Permaneceram', 'Sa√≠ram'], autopct='%1.1f%%', colors=['green', 'red'], startangle=90, explode=(0.1, 0))

# Adicionar t√≠tulo
plt.title('Taxa de Evas√£o de Clientes (Churn)', fontsize=14)

# Exibir o gr√°fico
plt.show()
```

```python
# Evas√£o por tempo de Contrato:
# agrupar tempo de contrato e evas√£o:
df_evasao_tempo_contrato = df.groupby(['tenure', 'churn']).size().unstack(fill_value=0)
# percentuual de evas√£o por tempo de contrato:
df_evasao_tempo_contrato['Cancelamentos (%)'] = (df_evasao_tempo_contrato[1] / df_evasao_tempo_contrato.sum(axis=1))*100
print(df_evasao_tempo_contrato[['Cancelamentos (%)']])
```

```python
fig = px.line(df_evasao_tempo_contrato.reset_index(),
              x='tenure',
              y='Cancelamentos (%)',
              markers=True,
              title = 'Taxa de evas√£o por meses de contrato')

fig.show()
```

A an√°lise da distribui√ß√£o de churn revela que aproximadamente 27% dos clientes evadiram (cancelaram seus contratos), enquanto 73% permaneceram. Isso indica que a taxa de churn √© um desafio relevante para a empresa.

Observou-se que a taxa de evas√£o √© mais alta entre clientes com menor tempo de contrato. A linha de evas√£o por tempo de contrato demonstra que a maior probabilidade de churn ocorre nos primeiros meses de contrato e tende a diminuir √† medida que o tempo de perman√™ncia aumenta.

## Contagem de Evas√£o por Vari√°veis Categ√≥ricas

- **G√™nero:** A distribui√ß√£o de churn por g√™nero n√£o apresentou diferen√ßas significativas, indicando que o g√™nero n√£o √© um fator determinante para a evas√£o.
- **Tipo de Contrato:** Clientes com contratos de curto prazo (M√™s a M√™s) apresentaram uma taxa de churn consideravelmente mais alta em compara√ß√£o com contratos anuais ou bienais.
- **M√©todo de Pagamento:** Alguns m√©todos de pagamento podem estar associados a taxas de churn mais altas. A an√°lise visual dos histogramas por m√©todo de pagamento e churn pode fornecer insights mais detalhados.

## Conclus√µes e Recomenda√ß√µes

A an√°lise aponta para os seguintes fatores como importantes contribuintes para a evas√£o de clientes na TelecomX:

1.  **Tempo de Contrato Curto:** Clientes nos primeiros meses de contrato s√£o mais propensos a cancelar.
2.  **Tipo de Contrato (M√™s a M√™s):** Este tipo de contrato est√° fortemente associado a uma maior taxa de churn.
3.  **Baixo Gasto Total:** Clientes que gastaram menos ao longo do tempo s√£o mais suscet√≠veis √† evas√£o.

**Recomenda√ß√µes:**

- **Foco na Reten√ß√£o nos Primeiros Meses:** Implementar estrat√©gias de engajamento e suporte proativo para novos clientes, especialmente nos primeiros meses, para aumentar a satisfa√ß√£o e reduzir a probabilidade de churn inicial.
- **Incentivar Contratos de Longo Prazo:** Oferecer benef√≠cios e descontos para clientes que optam por contratos anuais ou bienais para promover a fidelidade.
- **Identificar Clientes de Baixo Engajamento:** Monitorar o gasto e o uso de servi√ßos dos clientes para identificar aqueles com baixo engajamento e direcionar esfor√ßos de reten√ß√£o personalizados.
- **Analisar M√©todos de Pagamento:** Investigar os m√©todos de pagamento associados a taxas de churn mais altas para entender as raz√µes subjacentes e buscar solu√ß√µes.
